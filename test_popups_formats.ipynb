```python
import folium

print(folium.__file__)
print(folium.__version__)
```

```python
mapa = folium.Map(location=[45.372, -121.6972], zoom_start=12,
                  tiles='Stamen Terrain')

p0 = folium.Popup("Let's try quotes")
p1 = folium.Popup('and try "quotes"')
p2 = folium.Popup(u"Ça c'est chouette")
p3 = folium.Popup("Do scape sequences \n Work?")  # Nope and \r breaks the map!

text = u"""Can we do multiline?

Nope :-("""
p4 = folium.Popup(text)

dy = 0.01
folium.Marker([45.372, -121.6972-2*dy], popup=p0).add_to(mapa)
folium.Marker([45.372, -121.6972-dy], popup=p1).add_to(mapa)
folium.Marker([45.372, -121.6972], popup=p2).add_to(mapa)
folium.Marker([45.372, -121.6972+dy], popup=p3).add_to(mapa)
folium.Marker([45.372, -121.6972+2*dy], popup=p4).add_to(mapa)

mapa
```

# For anything fancier than simple text you need to use an IFrame

```python
from folium.element import IFrame

mapa = folium.Map(location=[45.372, -121.6972], zoom_start=12,
                  tiles='Stamen Terrain')


text = u"""
Let's try quotes<br>
and try "quotes"<br>
"Ça c'est chouette<br>
"""
# It is kind of anoying to set width and height for simple HTML text.
p = folium.Popup(IFrame(text, width=180, height=80))

folium.Marker([45.372, -121.6972], popup=p).add_to(mapa)

mapa
```
