```python
import folium

print(folium.__file__)
print(folium.__version__)
```

```python
lon = -(41 + 20/60. + 17/60/60.)
lat = -(20 + 59/60. + 35/60/60.)

min_lat = -25
max_lat = -15
min_lon = -45
max_lon = -35

mapa = folium.Map(location=[lat, lon], zoom_start=6,
                  min_lat=min_lat, max_lat=max_lat,
                  min_lon=min_lon, max_lon=max_lon)


kw = dict(marker_color="green", marker_icon="ok")

popup = '<b>{}</b>'.format('Upper Left Corner')
mapa.simple_marker(location=[max_lat, min_lon], popup=popup, **kw)

popup = '<b>{}</b>'.format('Lower Left Corner')
mapa.simple_marker(location=[min_lat, min_lon], popup=popup, **kw)

popup = '<b>{}</b>'.format('Lower Right Corner')
mapa.simple_marker(location=[min_lat, max_lon], popup=popup, **kw)

popup = '<b>{}</b>'.format('Upper Right Corner')
mapa.simple_marker(location=[max_lat, max_lon], popup=popup, **kw)

mapa
```
