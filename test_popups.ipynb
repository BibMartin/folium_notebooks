```python
import folium

print(folium.__file__)
print(folium.__version__)
```

```python
mapa = folium.Map(location=[45.372, -121.6972], zoom_start=12,tiles='Stamen Terrain')

p0 = folium.Popup(''.join(['Mt. Hood Meadows']*100), max_width=600)
p1 = folium.Popup(''.join(['Mt. Hood Meadows']*100), max_width=300)
p2 = folium.Popup(''.join(['Mt. Hood Meadows']*100), max_width=500)

m0 = folium.Marker([45.3288, -121.6625], popup=p0)
m1 = folium.Marker([45.3311, -121.7113], popup=p1)
m2 = folium.Marker([45.3300, -121.6823], popup=p2)

for m in [m0, m1, m2]:
    mapa.add_children(m)

mapa
```

```python
mapa = folium.Map(location=[45.372, -121.6972], zoom_start=12,tiles='Stamen Terrain')

m0 = folium.CircleMarker([45.3288, -121.6625], popup=p0)
m1 = folium.CircleMarker([45.3311, -121.7113], popup=p1)
m2 = folium.CircleMarker([45.3300, -121.6823], popup=p2)

for m in [m0, m1, m2]:
    mapa.add_children(m)

mapa
```

```python
mapa = folium.Map(location=[45.372, -121.6972], zoom_start=12,tiles='Stamen Terrain')

m0 = folium.PolyLine([[45.3288, -121.6625], [45.3311, -121.7113]], weight=8)
m0.add_children(p0)

m1 = folium.PolyLine([[45.3288+0.02, -121.6625], [45.3311+0.02, -121.7113]], weight=8)
m1.add_children(p1)

m2 = folium.PolyLine([[45.3288+0.04, -121.6625], [45.3311+0.04, -121.7113]], weight=8)
m2.add_children(p2)

for m in [m0, m1, m2]:
    mapa.add_children(m)

mapa
```
