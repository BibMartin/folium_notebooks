```python
import folium

print(folium.__file__)
print(folium.__version__)
```

```python
mapa = folium.Map(location=[45.372, -121.6972], zoom_start=12,tiles='Stamen Terrain')

mapa.simple_marker([45.3288, -121.6625], popup=''.join(['Mt. Hood Meadows']*100), popup_width=600)
mapa.simple_marker([45.3311, -121.7113], popup=''.join(['Mt. Hood Meadows']*100), popup_width=300)
mapa.simple_marker([45.3300, -121.6823], popup=''.join(['Mt. Hood Meadows']*100), popup_width=500)

mapa
```

```python
mapa = folium.Map(location=[45.372, -121.6972], zoom_start=12,tiles='Stamen Terrain')

mapa.circle_marker([45.3288, -121.6625], popup=''.join(['Mt. Hood Meadows']*100), popup_width=600)
mapa.circle_marker([45.3311, -121.7113], popup=''.join(['Mt. Hood Meadows']*100), popup_width=300)
mapa.circle_marker([45.3300, -121.6823], popup=''.join(['Mt. Hood Meadows']*100), popup_width=500)

mapa
```
