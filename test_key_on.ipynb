```python
import folium

print(folium.__file__)
print(folium.__version__)
```

```python
import json
import pandas as pd

with open('data/us-states.json') as f:
    states_json = json.load(f)

states = pd.Series({x['properties']['name']: float(k)
                    for k, x in enumerate(states_json['features'])})
```

```python
state_unemployment = 'data/US_Unemployment_Oct2012.csv'

df = pd.read_csv(state_unemployment)
df['State'] = states.index  # The data is wrong, but this is just an example.
```

```python
# Let Folium determine the scale.
mapa = folium.Map(location=[48, -102], zoom_start=3)

mapa.geo_json(geo_path='./data/us-states.json', data=df,
              columns=['State', 'Unemployment'],
              key_on='feature.properties.name',  # Check your GeoJSON for the name!
              fill_color='YlGn', fill_opacity=0.7, line_opacity=0.2,
              legend_name='Unemployment Rate (%)')

mapa.create_map('test.html')
mapa
```
