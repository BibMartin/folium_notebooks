```python
import folium

print(folium.__file__)
print(folium.__version__)
```

```python
import numpy as np

size = 1000

lons = np.random.randint(low=-180, high=180, size=size)
lats = np.random.randint(low=-90, high=90, size=size)
```

```python
mapa = folium.Map(location=[np.mean(lats), np.mean(lons)],
                  tiles='Cartodb Positron', zoom_start=1)

for k, (lon, lat) in enumerate(zip(lons, lats)):
    location = lat, lon
    popup = '<b>{}</b>'.format('station {}'.format(k))
    mapa.simple_marker(location=location, popup=popup,
                       clustered_marker=True)

mapa
```
