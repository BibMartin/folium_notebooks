```python
import folium

print(folium.__file__)
print(folium.__version__)
```

```python
mapa = folium.Map(location=[45.5236, -122.6750], zoom_start=13)

p = folium.RegularPolygonMarker(location=[45.5318, -122.6745],
                                popup='Broadway Bridge',
                                color='red',
                                opacity=1,
                                weight=2,
                                fill_color='blue',
                                fill_opacity=1,
                                number_of_sides=8,
                                rotation=0,
                                radius=10)
mapa.add_children(p)
mapa
```
