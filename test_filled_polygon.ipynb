```python
import folium

print(folium.__file__)
print(folium.__version__)
```

```python
mapa = folium.Map(location=[45.5236, -122.6750], zoom_start=13)

mapa.polygon_marker(location=[45.5318, -122.6745], popup='Broadway Bridge',
                    line_color='red',
                    fill_color='blue',
                    num_sides=8,
                    radius=10,
                    fill_opacity=1)
mapa
```

```python
from IPython.display import HTML

def inline_map(mapa):
    mapa._build_map()
    srcdoc = "data:text/html;base64," + mapa.HTML.encode('base64')
    embed = HTML('<iframe src="{}" '
                 'style="width: {}px; height: {}px; '
                 'border: none"></iframe>'.format(srcdoc, 700, 500))
    return embed

inline_map(mapa)
```
