```python
from folium import Map, FeatureGroup, Marker, Popup, LayerControl

mapa = Map(location=[45.372, -121.6972], zoom_start=12,
           tiles='Stamen Terrain')

feature_group = FeatureGroup(name='Some icons')
feature_group.add_children(Marker([45.3288, -121.6625], popup=Popup('Mt. Hood Meadows')))
feature_group.add_children(Marker([45.3311, -121.7113], popup=Popup('Timberline Lodge')))
mapa.add_children(feature_group)
mapa.add_children(LayerControl())

mapa
```
