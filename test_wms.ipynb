```python
import folium

print(folium.__file__)
print(folium.__version__)
```

```python
mapa = folium.Map(location=[44, -73], zoom_start=3)
wms_url = 'http://gis.srh.noaa.gov/arcgis/services/NDFDTemps/'
wms_url += 'MapServer/WMSServer'
wms_name = "Temperature"
wms_layers = 16
wms_format = "image/png"
mapa.add_wms_layer(wms_name=wms_name,
                   wms_url=wms_url,
                   wms_format=wms_format,
                   wms_layers=wms_layers,
                   wms_transparent=True)

mapa
```

### Leaflet's example

```python
from folium import plugins

mapa = folium.Map(location=[36.63, -69.50], zoom_start=5)


url = 'http://mesonet.agron.iastate.edu/cgi-bin/wms/nexrad/n0r.cgi'

folium.WmsTileLayer(name='nexrad',
                    url=url,
                    layers='nexrad-n0r-900913',
                    format='image/png').add_to(mapa)

folium.LayerControl().add_to(mapa)

mapa
```

### sci-wms is a Python WMS service for geospatial gridded data

```python
mapa = folium.Map(location=[28.1, -81.6], zoom_start=5)

url = 'http://sci-wms.whoi.edu/wms/datasets/usf_fvcom'
folium.WmsTileLayer(name='usf_fvcom',
                    url=url,
                    layers='u',
                    styles='pcolor_gnuplot2',
                    format='image/png').add_to(mapa)

folium.LayerControl().add_to(mapa)

mapa
```

```python
mapa = folium.Map(location=[28.1, -81.6], zoom_start=5)

url = 'http://sci-wms.whoi.edu/wms/datasets/usf_fvcom'
folium.WmsTileLayer(name='usf_fvcom',
                    url=url,
                    layers='v',
                    styles='pcolor_gnuplot2',
                    format='image/png').add_to(mapa)

folium.LayerControl().add_to(mapa)

mapa
```

```python
mapa = folium.Map(location=[28.1, -81.6], zoom_start=5)

url = 'http://sci-wms.whoi.edu/wms/datasets/usf_fvcom?VECTORSCALE=8.0'

folium.WmsTileLayer(name='usf_fvcom',
                    url=url,
                    layers='u,v',
                    styles='vectors_jet',
                    format='image/png').add_to(mapa)

folium.LayerControl().add_to(mapa)
mapa
```

### THREDDS also has a wms response

```python
mapa = folium.Map(location=[28.1, -81.6], zoom_start=4)

url = 'http://geoport-dev.whoi.edu/thredds/wms/coawst_4/use/fmrc/coawst_4_use_best.ncd?COLORSCALERANGE=33,37'

folium.WmsTileLayer(name='coawst_4',
                    url=url,
                    layers='salt',
                    styles='contour/occam',
                    format='image/png').add_to(mapa)
folium.LayerControl().add_to(mapa)
mapa
```

```python
mapa = folium.Map(location=[28.1, -81.6], zoom_start=4)

url = 'http://geoport-dev.whoi.edu/thredds/wms/coawst_4/use/fmrc/coawst_4_use_best.ncd'

folium.WmsTileLayer(name='coawst_4',
                    url=url,
                    layers='temp',
                    styles='boxfill/ferret',
                    format='image/png').add_to(mapa)
folium.LayerControl().add_to(mapa)
mapa
```

# Exploring the URL with OWSLib

```python
from owslib.wms import WebMapService

url = ('http://geoport-dev.whoi.edu/thredds/wms/coawst_4/use/fmrc/'
       'coawst_4_use_best.ncd')

service = WebMapService(url)
service.contents.keys()
```

```python
layer = service['salt']

layer.crsOptions, layer.boundingBoxWGS84, layer.title
```

```python
layer.styles
```

```python
layer.boundingBox, layer.boundingBoxWGS84
```

Can folium get the bounds of a wms layer?

```python
mapa = folium.Map(location=[36.63, -69.50], zoom_start=5)

folium.WmsTileLayer(name='coawst_4',
                    url=url,
                    layers='temp',
                    transparent=True,
                    format='image/png').add_to(mapa)

folium.LayerControl().add_to(mapa)
mapa.get_bounds()
```
